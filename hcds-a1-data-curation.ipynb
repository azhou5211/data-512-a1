# Data Collection and JSON output
Step 1: Collect data on Wikipedia traffic from 2008-2021 using two different API endpoints: Legacy Pagecounts API and Pageviews API.


```python
import json
import requests
```

### Setup endpoints

```python
# Endpoints for legacy and pageviews API
endpoint_legacy = 'https://wikimedia.org/api/rest_v1/metrics/legacy/pagecounts/aggregate/{project}/{access-site}/{granularity}/{start}/{end}'

endpoint_pageviews = 'https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/{project}/{access}/{agent}/{granularity}/{start}/{end}'
```

### Setup headers

```python
# Header for API call
headers = {
    'User-Agent': 'https://github.com/azhou5211',
    'From': 'ajz55@uw.edu'
}
```

### Setup API parameter settings

```python
# Legacy Desktop API settings
legacy_desktop = {"project" : "en.wikipedia.org",
                 "access-site" : "desktop-site",
                 "granularity" : "monthly",
                 "start" : "2008010100",
                # for end use 1st day of month following final month of data
                 "end" : "2022010100"
                    }

# Legacy Mobile API settings
legacy_mobile = {"project" : "en.wikipedia.org",
                 "access-site" : "mobile-site",
                 "granularity" : "monthly",
                 "start" : "2008010100",
                # for end use 1st day of month following final month of data
                 "end" : "2022010100"
                    }
```

```python
# Pageviews Desktop API settings
pageviews_desktop = {"project" : "en.wikipedia.org",
                    "access" : "desktop",
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2008010100",
                    # for end use 1st day of month following final month of data
                    "end" : '2022010100'
                        }

# Pageviews Mobile API settings
pageviews_mobile_web = {"project" : "en.wikipedia.org",
                    "access" : "mobile-web",
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2008010100",
                    # for end use 1st day of month following final month of data
                    "end" : '2022010100'
                        }

# Pageviews All API settings
pageviews_mobile_app = {"project" : "en.wikipedia.org",
                    "access" : "mobile-app",
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2008010100",
                    # for end use 1st day of month following final month of data
                    "end" : '2022010100'
                        }
```

### Create a function to call API

```python
# Function to call API
def api_call(endpoint, parameters):
    """
    Function used to call API. Will return response.
    
    :param endpoint: API URL endpoint
    :param parameters: Parameter settings in the API call
    """
    call = requests.get(endpoint.format(**parameters), headers=headers)
    response = call.json()
    return response
```

### Make API requests with endpoints and parameter settings

```python
# API Request for pagecounts desktop
pagecounts_desktop = api_call(endpoint_legacy, legacy_desktop)
pagecounts_desktop
```

```python
# API Request for pagecounts mobile
pagecounts_mobile = api_call(endpoint_legacy, legacy_mobile)
pagecounts_mobile
```

```python
# API Request for pageviews desktop
pageview_desktop = api_call(endpoint_pageviews, pageviews_desktop)
pageview_desktop
```

```python
# API Request for pageviews mobile_webb
pageview_mobile_web = api_call(endpoint_pageviews, pageviews_mobile_web)
pageview_mobile_web
```

```python
# API Request for pageviews mobile_app
pageview_mobile_app = api_call(endpoint_pageviews, pageviews_mobile_app)
pageview_mobile_app
```

# Data Processing
Step 2: Process the data for analysis.  
The cleaning process:  
- For data collected from the Pageviews API, combine the monthly values for mobile-app and mobile-web to create a total mobile traffic count for each month.
- For all data, separate the value of timestamp into four-digit year (YYYY) and two-digit month (MM) and discard values for day and hour (DDHH).

The end data table should have these columns: year, month, pagecount_all_views, pagecount_desktop_views, pagecount_mobile_views, pageview_all_views, pageview_desktop_views, pageview_mobile_views

```python
import pandas as pd
import numpy as np
```

### Create a dataframe of just year and month from 2008 to 2021
This will be the base dataframe

```python
# Get a list of year and month
date_range_list = pd.date_range('2008-01-01','2021-12-01', freq='MS').strftime("%Y-%m").tolist()
# Add it into a dataframe
df = pd.DataFrame({'time':date_range_list})
# Make year a seperate column
df['year'] = df['time'].apply(lambda s: s[:4])
# Make month a seperate column
df['month'] = df['time'].apply(lambda s: s[-2:])
# Get rid of the old time column
df.drop(columns=['time'], inplace=True)
df
```

### Convert all raw data JSON dictionaries into Pandas Dataframes

Below is a helper function used to convert raw json dictionary to dataframe. It also splits the timestamp column and converts it into a year and month column value. Only count, year, and month columns are leftover. All other columns will be dropped.

```python
# Helper function to convert raw json dictionary to pandas dataframe
def convert_dataframe(raw_data):
    """
    Converts a dictionary. It also splits the timestamp column into year and month columns.
    count, year and month columns will be left. All other columns will be dropped
    
    :param raw_data: dictionary of raw data
    :return: pandas dataframe of data
    """
    # Convert the dictionary into dataframe
    df = pd.json_normalize(raw_data, record_path=['items'])
    # Get the year from timestamp
    df['year'] = df['timestamp'].apply(lambda s: s[:4])
    # Get the month from timestamp
    df['month'] = df['timestamp'].apply(lambda s: s[4:6])
    # Keep only the last three columns (count, year, month)
    df = df.iloc[:,-3:]
    return df
```

```python
# Get the dataframe for pagecounts desktop
df_pagecounts_desktop = convert_dataframe(pagecounts_desktop)
df_pagecounts_desktop.head()
```

```python
# Get the dataframe for pagecounts mobile
df_pagecounts_mobile = convert_dataframe(pagecounts_mobile)
df_pagecounts_mobile.head()
```

```python
# Get the dataframe for pageview desktop
df_pageviews_desktop = convert_dataframe(pageview_desktop)
df_pageviews_desktop.head()
```

```python
# Get the dataframe for pageview mobile_web
df_pageviews_mobile_web = convert_dataframe(pageview_mobile_web)
df_pageviews_mobile_web.head()
```

```python
# Get the dataframe for pageview mobile_app
df_pageviews_mobile_app = convert_dataframe(pageview_mobile_app)
df_pageviews_mobile_app.head()
```

### Combine the dataframes into the base dataframe with year and month

```python
# Add in the pagecount desktop views into base df
df = df.merge(df_pagecounts_desktop, how="left").rename(columns={'count':'pagecount_desktop_views'})
df
```

```python
# Add in the pagecount mobile views into base df
df = df.merge(df_pagecounts_mobile, how="left").rename(columns={'count':'pagecount_mobile_views'})
df
```

```python
# Add in the pageview desktop views into base df
df = df.merge(df_pageviews_desktop, how="left").rename(columns={'views':'pageview_desktop_views'})
df
```

```python
# Add in the pageview mobile_web views into base df
df = df.merge(df_pageviews_mobile_web, how="left").rename(columns={'views':'pageview_mobile_web_views'})
df
```

```python
# Add in the pageview mobile_app views into base df
df = df.merge(df_pageviews_mobile_app, how="left").rename(columns={'views':'pageview_mobile_app_views'})
df
```

Make the nan values 0

```python
# Make sure the nan values are actually 0
df.fillna(0, inplace=True)
df
```

### Add up mobile_web and mobile_app to obtain a total mobile pageview column: pageview_mobile_views
Also drop the previous mobile_app and mobile_web columns

```python
# Add up the mobile_web and mobile_app to get total mobile pageviews
df['pageview_mobile_views'] = df['pageview_mobile_web_views'] + df['pageview_mobile_app_views']
# Delete the mobile_web and mobile_app columns as they are not needed
df.drop(columns=['pageview_mobile_web_views', 'pageview_mobile_app_views'], inplace=True)
df
```

### Combine pagecount desktop and mobile views for pagecount_all_views

```python
# Add up pagecount desktop and pagecount mobile for pagecount all
df['pagecount_all_views'] = df['pagecount_desktop_views'] + df['pagecount_mobile_views']
df
```

### Combine pageview desktop and mobile views for pageview_all_views

```python
# Add up pageview desktop and pageview mobile for pageview all
df['pageview_all_views'] = df['pageview_desktop_views'] + df['pageview_mobile_views']
df
```

Resort the columns into the order: year, month, pagecount_all_views, pagecount_desktop_views, pagecount_mobile_views, pageview_all_views, pageview_desktop_views, pageview_mobile_views

```python
# Reorder the dataframe columns
df = df[['year', 'month', 'pagecount_all_views', 'pagecount_desktop_views', 'pagecount_mobile_views', 'pageview_all_views', 'pageview_desktop_views', 'pageview_mobile_views']]
df
```

### Save the data into a csv

```python
# Save the dataframe as a csv
df.to_csv("en-wikipedia_traffic_200712-202108.csv", index=False)
```
